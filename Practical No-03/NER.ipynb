```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

piano_class_doc = nlp("Google announced its new Pixel smartphone in California on October 4, 2023. Sundar Pichai, the CEO, spoke about the innovative features. The event was streamed live on YouTube, attracting millions of viewers worldwide. The price starts at $799, and it will be available in stores in the United States by November. Pre-orders began on the official website shortly after the event. Samsung and Apple are expected to compete with their upcoming models. Industry analysts predict a strong holiday sales season for tech companies.")

displacy.serve(piano_class_doc, style="ent")
```
